## Data preparation code

```
import pandas as pd
import numpy as np
from collections import Counter
%matplotlib inline

df = pd.read_csv('data/input_data/250k_rndm_zinc_drugs_clean_3_canonized.csv')
print("Number of molecules in the ZINC dataset: %d" % df.shape[0])

# Split molecules into dataset A and B
A_indices = (df['qed'].values > 0.6) & (df['SAS'].values < 4)
df_A = df.iloc[A_indices]
df_B = df.iloc[~A_indices]
print("Number of molecules in dataset A: %d" % df_A.shape[0])
print("Number of molecules in dataset B: %d" % df_B.shape[0])
# Sanity check of datasets
print(len(df_A[(df_A['qed'].values < 0.6) & (df_A['SAS'].values > 4)])==0)
print(len(df_B[(df_B['qed'].values > 0.6) & (df_B['SAS'].values < 4)])==0)

# Split datasets into train and test datasets.

# split set A to train/test
test_sample_A = np.zeros(df_A.shape[0], dtype=bool)
test_sample_A[np.random.choice(df_A.shape[0], 6000, replace=False)] = True

df_test_A = df_A.iloc[test_sample_A]
df_train_A = df_A.iloc[~test_sample_A]
print("Number of molecules in dataset train_A: %d" % df_train_A.shape[0])
print("Number of molecules in dataset test_A: %d" % df_test_A.shape[0])

# split set B to train/test
test_sample_B = np.zeros(df_B.shape[0], dtype=bool)
test_sample_B[np.random.choice(df_B.shape[0], 6000, replace=False)] = True

df_test_B = df_B.iloc[test_sample_B]
df_train_B = df_B.iloc[~test_sample_B]
print("Number of molecules in dataset train_B: %d" % df_train_B.shape[0])
print("Number of molecules in dataset test_B: %d" % df_test_B.shape[0])

X_JTVAE = pd.read_csv('data/input_data/X_JTVAE_250k_rndm_zinc.csv')
X_JTVAE_zinc_train_A = X_JTVAE.loc[df_train_A.index]
X_JTVAE_zinc_test_A = X_JTVAE.loc[df_test_A.index]

X_JTVAE_zinc_train_B = X_JTVAE.loc[df_train_B.index]
X_JTVAE_zinc_test_B = X_JTVAE.loc[df_test_B.index]

print(X_JTVAE_zinc_train_A.shape, X_JTVAE_zinc_test_A.shape, 
      X_JTVAE_zinc_train_B.shape, X_JTVAE_zinc_test_B.shape)

X_JTVAE_zinc_train_A.to_csv('X_JTVAE_zinc_train_A.csv', index=False)
X_JTVAE_zinc_test_A.to_csv('X_JTVAE_zinc_test_A.csv', index=False)
X_JTVAE_zinc_train_B.to_csv('X_JTVAE_zinc_train_B.csv', index=False)
X_JTVAE_zinc_test_B.to_csv('X_JTVAE_zinc_test_B.csv', index=False)
```
